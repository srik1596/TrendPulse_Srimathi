### TASK 1 — Make the API Calls 

```python
# imported libraries
import requests
import json
import time
from datetime import datetime

top_stories = "https://hacker-news.firebaseio.com/v0/topstories.json" #given link

headers = {"User-Agent": "TrendPulse/1.0"}

response = requests.get(top_stories, headers=headers, timeout=5) #API Call

# Parse the JSON body into a Python dictionary - list of top story IDs
story_ids = response.json()

#print(len(story_ids))--default 500
print("List of Top 500 Stories:")
print(json.dumps(story_ids, indent=2))
```

```python
#given keywords
params_api = {
    'technology':['AI', 'software', 'tech', 'code', 'computer', 'data', 'cloud', 'API', 'GPU', 'LLM'],
    'worldnews':['war', 'government', 'country', 'president', 'election', 'climate', 'attack', 'global'],
    'sports':['NFL', 'NBA', 'FIFA', 'sport', 'game', 'team', 'player', 'league', 'championship'],
    'science':['research', 'study', 'space', 'physics', 'biology', 'discovery', 'NASA', 'genome'],
    'entertainment':['movie', 'film', 'music', 'Netflix', 'game', 'book', 'show', 'award', 'streaming']
    
}

final_data = {}   

all_posts = []

for category, keywords in params_api.items():
    print(f"\nProcessing: {category}")

    collected_data = []   # per category
    count = 0

    for story_id in story_ids:
        
        story = f"https://hacker-news.firebaseio.com/v0/item/{story_id}.json"

        try:
            response = requests.get(story, headers=headers, timeout=5)
            response.raise_for_status()
        except requests.exceptions.RequestException:
            print(f"Retrying story {story_id}...")
            try:
                response = requests.get(story, headers=headers, timeout=5)
                response.raise_for_status()
            except:
                print(f"Skipping story {story_id}")
                continue

        story_details = response.json()

        title = story_details["title"].lower()

        if any(keyword.lower() in title for keyword in keywords):

            if count >= 25:
                break
           #fetching story details
            post = {
                "post_id": story_details.get("id"),
                "title": story_details.get("title"),
                "category": category,   
                "score": story_details.get("score", 0),
                "num_comments": story_details.get("descendants", 0),
                "author": story_details.get("by", "unknown"),
                "collected_at": datetime.now().isoformat()
            }

            collected_data.append(post) #for category
            all_posts.append(post) #to know overall posts
            count += 1

    # store per category
    final_data[category] = collected_data

    time.sleep(2)

    print(f"Collected {count} posts")

print(f"Collected {len(all_posts)} post")
```

```python
import json

with open("trends_20260403.json", "w", encoding="utf-8") as f:
    json.dump(all_posts, f, indent=4)

print(f"Collected {len(all_posts)} Posts. Saved to trends_20260403.json")
```
