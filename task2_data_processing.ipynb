```python
#Load the JSON File
import json

with open("trends_20260403.json", "r", encoding="utf-8") as f:
    data=json.load(f)

print(f"Loaded {len(data)} stories from trends_20260403.json") #no. of rows loaded
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
df_copy = df.copy() #copy of df
```

```python
df.duplicated().sum()
```

```python
df['post_id'].duplicated().sum()
```

```python
# removing any rows with the same post_id
print("Before removing duplicates:", len(df))
df = df.drop_duplicates(subset='post_id')
print("After removing duplicates:", len(df))
```

```python
df.shape
```

```python
df.isnull().sum() #Missing values 
```

```python
#ensuring that score and num_comments datatypes are in integers
df.dtypes
```

```python
#Low quality — remove stories where score is less than 5
print("Before:", len(df))
df = df[df['score']>=5] #filtered score 
print("After:", len(df))
```

```python
#Whitespace — strip extra spaces from the title column
df['title'] = df['title'].str.strip()
```

```python
df.shape
```

```python
df.to_csv("trends_cleaned.csv")

print("Saved 88 rows to trends_cleaned.csv")
```
