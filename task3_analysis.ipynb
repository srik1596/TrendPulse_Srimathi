```python
import pandas as pd

df = pd.read_csv("trends_cleaned.csv")

print("First 5 rows:")
df.head(5) #1st 5 rows
```

```python
print("Loaded data:", df.shape #no. of rows and columns
```

```python
print("Average score:", df['score'].mean().round(2))
```

```python
print("Average comments:", df['num_comments'].mean().round(2))
```

```python
#mean, median, and standard deviation of score
import numpy as np

scores = df['score'].to_numpy() #converted this column to numpy array
print("--- NumPy Stats ---")
print("Mean score   :", np.mean(scores).round(2))
print("Median score :", np.median(scores).round(2))
print("Std deviation:", np.std(scores).round(2))
print("Max score    :", np.max(scores))  #highest score
print("Min score    :", np.min(scores))  #lowest score
```

```python
print("Most stories in:", df['category'].value_counts().idxmax())
```

```python
#Which story has the most comments? Print its title and comment count.
most_commented_index = df.loc[df["num_comments"].idxmax()]

print(f"Most commented story: {most_commented_index["title"]} - {most_commented_index["num_comments"]} comments")
```

```python
#adding new columns
df['engagement'] = df["num_comments"] / (df['score']+1)

avg_score = df['score'].mean().round(2)

df['is_popular'] = df['score'] > avg_score
```

```python
df.to_csv("trends_analysed.csv", index=False)
print("Saved to trends_analysed.csv")
```
